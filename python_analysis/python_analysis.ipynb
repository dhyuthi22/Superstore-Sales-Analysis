```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_excel('/content/Sample - Superstore.xlsx')

# Basic info
print(df.head())
print(df.info())
print(df.describe())
df.columns

# Clean data
df.drop_duplicates(inplace=True)
print(df.isnull().sum())
```

```python
# Change format of date columns to date format

date_columns = ['Order Date', 'Ship Date']
for col in date_columns:
    df[col] = pd.to_datetime(df[col], errors='coerce')

print(df[date_columns].isnull().sum())
print(df.head())
```

Exploratory Data Analysis

```python
state_counts = df['State'].value_counts().reset_index()
state_counts.columns = ['state', 'count']

print(state_counts.head(10))

plt.figure(figsize=(12, 6))
plt.bar(state_counts['state'][:10], state_counts['count'][:10], color='darkblue')

plt.title('Top 10 States by Number of Orders (20142017)', fontsize=14)
plt.xlabel('State')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

From above analysis, we can say that California is the state where most Orders were placed which is 2001 in the entire time period starting from 2014 - 2017. Then it is New York at 1128 and Texas at 985.

```python
# Data for the year 2017
df_2017 = df[df['Order Date'].dt.year == 2017]

# Count orders by state for 2017
state_counts_2017 = df_2017['State'].value_counts().reset_index()
state_counts_2017.columns = ['State', 'Order Count']

# Print top 10 states
print(state_counts_2017.head(10))

plt.figure(figsize=(12, 6))
plt.bar(state_counts_2017['State'][:10], state_counts_2017['Order Count'][:10], color='darkblue')

plt.title('Top 10 States by Number of Orders in 2017', fontsize=14)
plt.xlabel('State')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

From the above, we can see that the orders placed in California is the largest.

```python
ca_2017 = df[(df['Order Date'].dt.year == 2017) & (df['State'] == 'California')]

# Count orders by Sub-Category
subcat_counts = ca_2017['Sub-Category'].value_counts().reset_index()
subcat_counts.columns = ['Sub-Category', 'Order Count']

print(subcat_counts)

# Plot top sub-categories
plt.figure(figsize=(12, 6))
plt.bar(subcat_counts['Sub-Category'], subcat_counts['Order Count'], color='darkblue')
plt.title('Orders by Sub-Category in California (2017)', fontsize=14)
plt.xlabel('Sub-Category')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

In 2017, Binders were the most purchased product in California, followed by Paper, Art, etc.

```python
ca_office_supplies = df[
    (df['Order Date'].dt.year == 2017) &
    (df['State'] == 'California') &
    (df['Category'] == 'Office Supplies')
]

# Aggregate order counts and average discounts by Sub-Category
summary = ca_office_supplies.groupby('Sub-Category').agg(
    order_count=('Sub-Category', 'count'),
    avg_discount=('Discount', 'mean')
).reset_index()

# Sort by order_count descending
summary = summary.sort_values(by='order_count', ascending=False)
print(summary)

# Plot order counts
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.bar(summary['Sub-Category'], summary['order_count'], color='royalblue')
plt.title('Orders by Office Supplies Sub-Category in CA (2017)')
plt.xlabel('Sub-Category')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45, ha='right')

# Plot average discount
plt.subplot(1, 2, 2)
plt.bar(summary['Sub-Category'], summary['avg_discount'], color='coral')
plt.title('Average Discount by Sub-Category in CA (2017)')
plt.xlabel('Sub-Category')
plt.ylabel('Average Discount')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

The above data says that relatively little product discounting was done in California. People bought regardless of the reductions as a result.

```python
df_filtered = df[(df['Order Date'].dt.year >= 2014) & (df['Order Date'].dt.year <= 2017)]

# Group by State and sum Sales
sales_by_state = df_filtered.groupby('State')['Sales'].sum().reset_index()

# Sort descending by total sales
top_sales_states = sales_by_state.sort_values(by='Sales', ascending=False).head(10)
top_sales_states.columns = ['State', 'Total_sales']

# Print top 10 states by sales
print(top_sales_states)

# Plotting
plt.figure(figsize=(12, 6))
plt.bar(top_sales_states['State'], top_sales_states['Total_sales'], color='darkblue')

plt.title('Top 10 States by Total Sales (20142017)', fontsize=14)
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

From above plot and value of total sales, we can see that it is similar to count of products sold.
California has the most sales with 457687.6315, then New York with 310876.2710 and Texas with 170188.0458 and so on.

```python
# California Monthly Sales(2014-2017)

df_ca = df[df['State'] == 'California']

# Set Order Date as index for time series plotting
df_ca.set_index('Order Date', inplace=True)

# Resample monthly and sum sales
monthly_sales = df_ca['Sales'].resample('M').sum()

# Print the monthly sales values
print("Monthly Sales in California (20142017):")
print(monthly_sales)

# Plotting the time series
plt.figure(figsize=(14, 6))
plt.plot(monthly_sales.index, monthly_sales.values, color='mediumblue', linewidth=2, marker='o')

plt.title('Monthly Sales Trend in California (20142017)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Total Sales')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# California Monthly Profit(2014-2017)

df_ca = df[df['State'] == 'California']

# Set 'Order Date' as index for time series
df_ca.set_index('Order Date', inplace=True)

# Resample by month and sum the Profit
monthly_profit = df_ca['Profit'].resample('M').sum()

# Print monthly profit values
print("Monthly Profit in California (20142017):")
print(monthly_profit)

plt.figure(figsize=(14, 6))
plt.plot(monthly_profit.index, monthly_profit.values, color='darkgreen', linewidth=2, marker='o')

plt.title('Monthly Profit Trend in California (20142017)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Total Profit')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# Count of Products bought based on Region

region_orders = df_filtered['Region'].value_counts().reset_index()
region_orders.columns = ['Region', 'Order Count']

print("Orders placed per Region (20142017):")
print(region_orders)

# Plotting region-wise order counts
plt.figure(figsize=(10, 5))
plt.bar(region_orders['Region'], region_orders['Order Count'], color='steelblue')

plt.title('Orders Placed per Region (20142017)', fontsize=14)
plt.xlabel('Region')
plt.ylabel('Number of Orders')
plt.xticks(rotation=0)
plt.tight_layout()

plt.show()
```
